# Variance

The test error is called Variance

When selecting random training data points, the test set has different errors when predicting. When the error difference is very large in the test dataset from one dataset to another that is called High Variance. High error of a model means the model is overfit.

Overfit = high variance

# Bias

The traning error is called Bias

Bias is actually the measurement of how accurately the model can capture a pattern in the training dataset.

Underfit = high bias

We want a model with low bias and low variance

Ways to get balanced fit model
1. Cross validation
2. Regularization
3. Dimensionality Reduction
4. Ensemble Techniques
