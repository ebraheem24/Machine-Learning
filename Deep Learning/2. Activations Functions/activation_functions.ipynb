### **Implementation of activation functions in python**

##### **Sigmoid**

It brings the values between 0 and 1

Use sigmoid in output layer and in all other places try to use tanh.

```python
import math 

def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(1)
```

```python
sigmoid(100)
```

```python
sigmoid(-50)
```

```python
sigmoid(0.5)
```

##### **tanh**

It brings the values between -1 and 1

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(7)
```

```python
tanh(0.5)
```

```python
tanh(-10)
```

##### **ReLU**

If the value is greater than 0, this fucntion will return the same value and if the values is lesser than 0 it will return 0 for all values.

For hidden layers, if you are not sure which activation function to use then always use ReLU function as your default choice.

```python
def relu(x):
    return max(0,x)
```

```python
relu(-10)
```

```python
relu(10)
```

##### **Leaky ReLU**

If the value is greater than 0, this fucntion will return the same value and if the values is lesser than 0 it will mulitply the values with 0.1 and return the value.

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(10)
```

```python
leaky_relu(-1)
```
